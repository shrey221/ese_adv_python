```python
data=pd.read_csv('scraped_final.csv')
data
```

# Resolving the title section  

```python
for index, title in enumerate(data['title']):
    if "." in (title.split()[0]):
        data['title'][index] = title[2:]
    else:
        pass 
```

# Resolving the Rating Section 

```python
data
```

```python

for i in range(len(data['rating'])):
    data['rating'][i] = data['rating'][i].split()[0]
```

```python
data
```

```python
for i in range(len(data['duration'])):
    parts = data['duration'][i].split() 
    if len(parts) == 2:  
        hours = int(parts[0][:-1]) 
        minutes = int(parts[1][:-1])  
        total_minutes = hours * 60 + minutes  
    elif 'h' in data['duration'][i]:  
        hours = int(parts[0][:-1])  
        total_minutes = hours * 60  
    elif 'm' in data['duration'][i]:  
        minutes = int(parts[0][:-1]) 
        total_minutes = minutes  
    else:
        total_minutes = 0  
    data['duration'][i] = total_minutes  
```
